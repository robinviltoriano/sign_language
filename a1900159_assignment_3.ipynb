### General Function for Modelling

This section containts functions that are used in the later part of the assignment

```python
def plot_loss_accuracy(history, model_name=''):
    """
    Plots the loss and accuracy curves for a given training history.

    Parameters:
    history (keras.callbacks.History): The training history object obtained from model.fit().

    Returns:
    None
    """
    # Plot the loss and accuracy curves
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))

    axs[0].plot(history.history['loss'], label='train', marker='o')
    axs[0].plot(history.history['val_loss'], label='validation', marker='o')
    axs[0].set_title('Loss')
    axs[0].set_xlabel('Epoch')

    axs[1].plot(history.history['accuracy'], label='train', marker='o')
    axs[1].plot(history.history['val_accuracy'], label='validation', marker='o')
    axs[1].set_title('Accuracy')
    axs[1].set_xlabel('Epoch')

    axs[1].legend(bbox_to_anchor=(-0.1, -0.1), loc='upper center', ncol=2, fontsize=13, title_fontsize=13)

    fig.suptitle(f'Loss and Accuracy Curves - {model_name}')
    plt.show()

def plot_cm(y_true, y_pred, model_name='', pplot=True):
    """
    Plot the confusion matrix for a given set of true labels and predicted labels.

    Parameters:
    - y_true (array-like): True labels.
    - y_pred (array-like): Predicted labels.
    - model_name (str, optional): Name of the model. Default is an empty string.
    - pplot (bool, optional): Whether to plot the confusion matrix. Default is True.

    Returns:
    - cm_nn (array-like): The confusion matrix.
    """
    # Get the confusion matrix
    y_pred = np.argmax(y_pred, axis=1)
    cm_nn = confusion_matrix(np.argmax(y_true, axis=1), y_pred)

    # Plot the confusion matrix
    if pplot:
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm_nn, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix - {model_name}')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.xticks(ticks=np.arange(24)+0.5, labels=letters_no_jz)
        plt.yticks(ticks=np.arange(24)+0.5, labels=letters_no_jz)
        plt.show()

    return cm_nn

def calculate_accuracy(y_true, y_pred):
    """
    Calculates the per-class accuracy and median accuracy given the true labels and predicted labels.

    Parameters:
    - y_true (numpy.ndarray): True labels of the samples.
    - y_pred (numpy.ndarray): Predicted labels of the samples.

    Returns:
    - accuracy_dict (dict): A dictionary containing the per-class accuracy and median accuracy.
                           The keys are the class labels and 'median', and the values are the corresponding accuracies.
    """

    # Calculate per-class accuracy
    predicted_labels = np.argmax(y_pred, axis=1)
    true_labels = np.argmax(y_true, axis=1)

    # Calculate per-class accuracy
    accuracy_dict = {}
    for i in range(24):
        class_indices = np.where(true_labels == i)
        class_accuracy = accuracy_score(true_labels[class_indices], predicted_labels[class_indices])
        accuracy_dict[letters_no_jz[i]] = class_accuracy

    # Calculate median accuracy
    median_accuracy = np.median([x for x in accuracy_dict.values()])
    accuracy_dict['median'] = median_accuracy
    accuracy_dict['mean'] = accuracy_score(true_labels, predicted_labels)

    return accuracy_dict

def evaluate_model(y_true, y_pred, history=None, model_name='', plot_cm_flag=True, plot_loss_accuracy_flag=True, show_all_accuracy=False):
    """
    Evaluates the performance of a model by calculating various accuracy metrics and generating visualizations.

    Parameters:
    - y_true (array-like): The true labels.
    - y_pred (array-like): The predicted labels.
    - history (optional, object): The training history of the model. Default is None.
    - model_name (optional, str): The name of the model. Default is an empty string.
    - plot_cm_flag (optional, bool): Flag to indicate whether to plot the confusion matrix. Default is True.
    - plot_loss_accuracy_flag (optional, bool): Flag to indicate whether to plot the loss and accuracy curves. Default is True.
    - show_all_accuracy (optional, bool): Flag to indicate whether to show the accuracy for each letter. Default is False.

    Returns:
    - accuracy_dict (dict): A dictionary containing the accuracy metrics.

    """
    # Plot the loss and accuracy curves
    if plot_loss_accuracy_flag:
        plot_loss_accuracy(history, model_name)

    # Calculate the accuracy metrics
    accuracy_dict = calculate_accuracy(y_true, y_pred)

    # Get the Best Accuracy Letter
    max_key = max(accuracy_dict, key=accuracy_dict.get)
    max_value = accuracy_dict[max_key]
    print(f"Best Letter Accuracy: {max_key} - {round(max_value*100,2)}%")

    # Get the Worst Accuracy Letter
    min_key = min(accuracy_dict, key=accuracy_dict.get)
    min_value = accuracy_dict[min_key]
    print(f"Worst Letter Accuracy: {min_key} - {round(min_value*100,2)}%\n")

    # Get the letters with accuracy below 85%
    letter_below_85 = {k:v for k,v in accuracy_dict.items() if (v < 0.85) and (k != 'median') and (k != 'mean')}
    print(f"{len(letter_below_85)} Letters with Accuracy Below 85%:")
    for k,v in letter_below_85.items():
        print(f"{k} - {round(v*100,2)}%")

    print('')
    # Print the overall accuracy
    print(f"Overall Median Accuracy: {round(accuracy_dict['median']*100,2)}%")
    print(f"Overall Mean Accuracy: {round(accuracy_dict['mean']*100,2)}%\n")

    # Print the per-letter accuracy
    if show_all_accuracy:
        print("Per-Letter Accuracy:")
        for k,v in accuracy_dict.items():
            if k != 'median' and k != 'mean':
                print(f"{k} - {round(v*100,2)}%")

    # Get the confusion matrix
    cm = plot_cm(y_true, y_pred, model_name, pplot=plot_cm_flag)

    # Find the three most common errors
    np.fill_diagonal(cm, 0)  # Ignore correct classifications
    errors = np.unravel_index(np.argsort(-cm, axis=None), cm.shape)
    common_errors = list(zip(errors[0][:3], errors[1][:3], cm[errors[0][:3], errors[1][:3]]))

    print('')

    # Print the three most common errors
    print("Three most common errors (predicted label, count | true label):")
    for true_label, pred_label, count in common_errors:
        print(f"Predicted label: {letters_no_jz[pred_label]}, Count: {count} | True label: {letters_no_jz[true_label]}")

    return accuracy_dict


def exponential_decay(lr0, s):
    """
    Returns a function that computes the learning rate using exponential decay.

    Parameters:
    - lr0 (float): The initial learning rate.
    - s (float): The decay rate.

    Returns:
    - exponential_decay_fn (function): A function that takes an epoch as input and returns the learning rate.
    """
    
    # Define the exponential decay function
    def exponential_decay_fn(epoch):
        return lr0 * 0.1**(epoch / s)
    return exponential_decay_fn

def OneCycleSchedule(epoch, max_lr, epochs, steps_per_epoch):
    """
    Generates a learning rate function for the One Cycle Policy.

    Args:
        epoch (int): The current epoch.
        max_lr (float): The maximum learning rate.
        epochs (int): The total number of epochs.
        steps_per_epoch (int): The number of steps per epoch.

    Returns:
        float: The learning rate for the given step.

    """
    
    # Calculate the total number of steps
    def lr_fn(step):
        # Calculate the phase
        total_steps = epochs * steps_per_epoch
        step += epoch * steps_per_epoch
        phase = step / total_steps

        if phase < 0.3:
            lr = max_lr * (phase / 0.3)
        elif phase < 0.7:
            lr = max_lr
        else:
            lr = max_lr * (1 - (phase - 0.7) / 0.3)

        return lr

    return lr_fn

# create list of letters without J and Z
letters_no_jz = np.delete(letters.copy(), [9, 25])
```

### Neural Network

```python
# Create a partial function for the Dense layer
DenseLayer = partial(keras.layers.Dense, activation='elu', kernel_initializer='he_normal')

# Create a function to create the neural network model
def create_nn_model(input_shape, layers, optimizer, learningrate):
    """
    Create a neural network model with the specified input shape, layers, optimizer, and learning rate.

    Parameters:
    input_shape (tuple): The shape of the input data.
    layers (list): A list of integers representing the number of neurons in each hidden layer.
    optimizer (keras.optimizers.Optimizer): The optimizer to use for training the model.
    learningrate (float): The learning rate for the optimizer.

    Returns:
    keras.models.Sequential: The created neural network model.
    """
    # Create the model
    model = keras.models.Sequential()
    
    # Add the input layer
    model.add(keras.layers.Input(shape=input_shape))
    
    # Add the hidden layers
    for layer in layers:
        model.add(DenseLayer(layer))
        
    # Add the output layer
    model.add(keras.layers.Dense(24, activation='softmax', kernel_initializer='glorot_uniform'))

    # Compile the model
    optimizer = optimizer(learning_rate=learningrate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    return model
```

```python
# Set the input shape
input_shape = X_train.shape[1:]

# Define the layers
layers = [128, 128, 64, 64, 32]

# Create the neural network model
model_nn = create_nn_model(input_shape, layers, optimizer=keras.optimizers.legacy.Adam, learningrate=0.001)

# Display the model summary
model_nn.summary()
```

```python
# Train the model
history_nn = model_nn.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val), verbose=0)
```

```python
# Evaluate the model
y_val_pred = model_nn.predict(X_val)
eval_nn_base = evaluate_model(y_val, y_val_pred, history_nn, 'Neural Network', plot_cm_flag=False)
```

### Convolutional Neural Network

```python
# prepare the data for CNN
X_train_cnn = X_train.reshape(-1, 32, 32, 1)
X_val_cnn = X_val.reshape(-1, 32, 32, 1)
```

```python
Conv2DLayer = partial(keras.layers.Conv2D, activation='elu', kernel_initializer='he_normal')

# Create the CNN model
model_cnn = keras.models.Sequential([
    keras.layers.Input(shape=(32, 32, 1)),
    Conv2DLayer(32, (3, 3)),
    keras.layers.MaxPooling2D((2, 2)),
    Conv2DLayer(64, (3, 3)),
    keras.layers.MaxPooling2D((2, 2)),
    Conv2DLayer(64, (3, 3)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='elu', kernel_initializer='he_normal'),
    keras.layers.Dense(24, activation='softmax')
])

# Compile the model
model_cnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the model summary
model_cnn.summary()
```

```python
# Train the model
history_cnn = model_cnn.fit(X_train_cnn, y_train, epochs=20, batch_size=32, validation_data=(X_val_cnn, y_val), verbose=0)
```

```python
# Evaluate the model
y_val_pred_cnn = model_cnn.predict(X_val_cnn)
eval_cnn_base = evaluate_model(y_val, y_val_pred_cnn, history_cnn, 'Convolutional Neural Network', plot_cm_flag=False)
```

```python
layers_names = [f'{i}' for i in range(1,len(layers)+1)]

# Get the index of the maximum accuracy
max_index = evaluation_nn_layers.index(max(evaluation_nn_layers))

# Plot the accuracy vs number of layers
plt.plot(layers_names, evaluation_nn_layers)
plt.plot(layers_names, evaluation_nn_layers,'o')
plt.plot(layers_names[max_index], evaluation_nn_layers[max_index],'go')
plt.title('Accuracy vs Number of Layers')
plt.xlabel('Number of Layers')
plt.ylabel('Val Accuracy')
plt.show()
```

Just using the first four layers gives the highest overall accuracy, thus only four layers will be used in the next step

### Optimiser

```python
# Define the optimizers
optimizers = [keras.optimizers.legacy.SGD, keras.optimizers.legacy.Adam, keras.optimizers.legacy.RMSprop, keras.optimizers.legacy.Nadam]
evaluation_nn_opt = []

# Evaluate the model with different optimizers
for opt in tqdm(optimizers):
  model_nn = create_nn_model(input_shape=input_shape, layers=layers[:4], optimizer=opt, learningrate=learningrate)
  history = model_nn.fit(X_train, y_train, epochs=epoch, batch_size=batch_size,
                         validation_data=(X_val, y_val), callbacks=[early_stopping], verbose=0)
  max_val_acc = np.max(history.history['val_accuracy'])
  evaluation_nn_opt.append(max_val_acc)
```

```python
optimizer_names = ['SGD', 'Adam', 'RMSprop', 'Nadam']

# Get the index of the maximum accuracy
max_index = evaluation_nn_opt.index(max(evaluation_nn_opt))

# Plot the accuracy vs optimizer
plt.bar(x=optimizer_names, height=evaluation_nn_opt)
plt.bar(x=optimizer_names[max_index], height=evaluation_nn_opt[max_index], color='g')
plt.title('Accuracy vs Optimizer')
plt.xlabel('Optimizer')
plt.ylabel('Val Accuracy')
plt.show()
```

### CNN Model 2 with Residual Connection

In this section, the CNN model will be trained using residual connection. We want to see if the model performance can be improved by using residual connection.

```python
# This code was taken from MyUni page Week 11

# Define the Residual Unit Class
class ResidualUnit(keras.layers.Layer):
    def __init__ (self, filters, strides=1, activation="elu", kernel_initializer="he_normal", **kwargs):
        
        # Call the parent class constructor
        super(). __init__ (**kwargs)
        
        # Get the activation function
        self.activation = keras.activations.get(activation)

        # Define the main layers
        self.main_layers = [

            keras.layers.Conv2D(filters, 3, strides=strides, padding="same", kernel_initializer=kernel_initializer, use_bias=False),
            keras.layers.BatchNormalization(),
            self.activation,

            keras.layers.Conv2D(filters, 3, strides=1, padding="same", kernel_initializer=kernel_initializer, use_bias=False),
            keras.layers.BatchNormalization()

            ]

        # Define the skip layers
        self.skip_layers = []
        if strides > 1:
            self.skip_layers = [

                keras.layers.Conv2D(filters, 1, strides=strides, padding="same", kernel_initializer=kernel_initializer, use_bias=False),
                keras.layers.BatchNormalization()
                ]

    # Define the call method
    def call(self, inputs):
        # Get the input value
        Z = inputs
        
        # Pass the input through the main layers
        for layer in self.main_layers:
            Z = layer(Z)
            
        # Define the skip connection input
        skip_Z = inputs
        
        # Pass the skip connection input through the skip layers
        for layer in self.skip_layers:
            skip_Z = layer(skip_Z)
        return self.activation(Z + skip_Z)
```

```python
# Define the CNN Residual Model
def create_cnn_residual_model(input_shape, initial_filters, residual_layers, actfn, optimizer, learningrate, kernel_initializer, kernel_regularizer):
    # Create the model
    model = keras.models.Sequential()
    
    # Add the input layer
    model.add(keras.layers.Input(shape=input_shape))

    # Add the initial layers
    model.add(keras.layers.Conv2D(initial_filters, 5, padding="same", kernel_initializer=kernel_initializer, use_bias=False))
    model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.Activation(actfn))
    model.add(keras.layers.MaxPool2D(pool_size=3, strides=2, padding="same"))

    # Add the residual layers
    prev_filters = initial_filters
    for filters in residual_layers:
        if filters == prev_filters:
            strides = 1
        else:
            strides = 2
        model.add(ResidualUnit(filters, strides=strides, kernel_initializer=kernel_initializer))
        prev_filters = filters

    # Add the final layers
    model.add(keras.layers.MaxPool2D(pool_size=2, padding="same"))
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(128, activation=actfn, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer))
    model.add(keras.layers.Dropout(rate=0.25))
    model.add(keras.layers.Dense(64, activation=actfn, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer))
    model.add(keras.layers.Dropout(rate=0.25))
    model.add(keras.layers.Dense(32, activation=actfn, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer))
    model.add(keras.layers.Dense(24, activation="softmax", kernel_initializer="glorot_uniform"))

    # Compile the model
    model.compile(loss="categorical_crossentropy", optimizer=optimizer(learning_rate=learningrate), metrics=["accuracy"])

    return model
```
