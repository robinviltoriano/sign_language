## Data Preparation

```python
# Split the data into X and y
X_train = df_train.drop('label', axis=1).values
y_train = df_train['label'].values

X_val = df_val.drop('label', axis=1).values
y_val = df_val['label'].values

X_test = df_test.drop('label', axis=1).values
y_test = df_test['label'].values

# Normalize the data
X_train = X_train / 255
X_val = X_val / 255
X_test = X_test / 255

# Change the target variable to categorical
lb=LabelBinarizer()
y_train=lb.fit_transform(y_train)
y_val=lb.fit_transform(y_val)
y_test=lb.fit_transform(y_test)

print(f"X Train Shape: {X_train.shape} | y Train Shape: {y_train.shape}")
print(f"X Validation Shape: {X_val.shape} | y Validation Shape: {y_val.shape}")
print(f"X Test Shape: {X_test.shape} | y Test Shape: {y_test.shape}")
```

### General Function for Modelling

```python
def plot_loss_accuracy(history, model_name=''):
    """
    Plots the loss and accuracy curves for a given training history.

    Parameters:
    history (keras.callbacks.History): The training history object obtained from model.fit().

    Returns:
    None
    """
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    
    axs[0].plot(history.history['loss'], label='train', marker='o')
    axs[0].plot(history.history['val_loss'], label='validation', marker='o')
    axs[0].set_title('Loss')
    axs[0].set_xlabel('Epoch')

    axs[1].plot(history.history['accuracy'], label='train', marker='o')
    axs[1].plot(history.history['val_accuracy'], label='validation', marker='o')
    axs[1].set_title('Accuracy')
    axs[1].set_xlabel('Epoch')
    
    axs[1].legend(bbox_to_anchor=(-0.1, -0.1), loc='upper center', ncol=2, fontsize=13, title_fontsize=13)

    fig.suptitle(f'Loss and Accuracy Curves - {model_name}')
    plt.show()
    
def plot_cm(y_true, y_pred, model_name='', pplot=True):
    """
    Plot the confusion matrix for a given set of true labels and predicted labels.

    Parameters:
    - y_true (array-like): True labels.
    - y_pred (array-like): Predicted labels.
    - model_name (str, optional): Name of the model. Default is an empty string.
    - pplot (bool, optional): Whether to plot the confusion matrix. Default is True.

    Returns:
    - cm_nn (array-like): The confusion matrix.
    """

    y_pred = np.argmax(y_pred, axis=1)
    cm_nn = confusion_matrix(np.argmax(y_true, axis=1), y_pred)

    if pplot:
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm_nn, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix - {model_name}')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.xticks(ticks=np.arange(24)+0.5, labels=letters_no_jz)
        plt.yticks(ticks=np.arange(24)+0.5, labels=letters_no_jz)
        plt.show()
    
    return cm_nn

def calculate_accuracy(y_true, y_pred):
    """
    Calculates the per-class accuracy and median accuracy given the true labels and predicted labels.

    Parameters:
    - y_true (numpy.ndarray): True labels of the samples.
    - y_pred (numpy.ndarray): Predicted labels of the samples.

    Returns:
    - accuracy_dict (dict): A dictionary containing the per-class accuracy and median accuracy.
                           The keys are the class labels and 'median', and the values are the corresponding accuracies.
    """

    # Calculate per-class accuracy
    predicted_labels = np.argmax(y_pred, axis=1)
    true_labels = np.argmax(y_true, axis=1)

    accuracy_dict = {}
    for i in range(24):
        class_indices = np.where(true_labels == i)
        class_accuracy = accuracy_score(true_labels[class_indices], predicted_labels[class_indices])
        accuracy_dict[letters_no_jz[i]] = class_accuracy

    # Calculate median accuracy
    median_accuracy = np.median([x for x in accuracy_dict.values()])
    accuracy_dict['median'] = median_accuracy
    
    return accuracy_dict

def evaluate_model(y_true, y_pred, history=None, model_name='', plot_cm_flag=True, plot_loss_accuracy_flag=True, show_all_accuracy=False):
    
    if plot_loss_accuracy_flag:
        plot_loss_accuracy(history, model_name)
        
    accuracy_dict = calculate_accuracy(y_true, y_pred)
    
    # Get the Best Accuracy Letter
    max_key = max(accuracy_dict, key=accuracy_dict.get)
    max_value = accuracy_dict[max_key]
    print(f"Best Letter Accuracy: {max_key} - {round(max_value*100,2)}%")
    
    # Get the Worst Accuracy Letter
    min_key = min(accuracy_dict, key=accuracy_dict.get)
    min_value = accuracy_dict[min_key]
    print(f"Worst Letter Accuracy: {min_key} - {round(min_value*100,2)}%\n")
    
    letter_below_85 = {k:v for k,v in accuracy_dict.items() if (v < 0.85) and (k != 'median')}
    print(f"{len(letter_below_85)} Letters with Accuracy Below 85%:")
    for k,v in letter_below_85.items():
        print(f"{k} - {round(v*100,2)}%")
    
    print('')
    
    print(f"Median Accuracy: {round(accuracy_dict['median']*100,2)}%\n")
    
    if show_all_accuracy:
        print("Per-Letter Accuracy:")
        for k,v in accuracy_dict.items():
            if k != 'median':
                print(f"{k} - {round(v*100,2)}%")
    
    # get the confusion matrix
    cm = plot_cm(y_true, y_pred, model_name, pplot=plot_cm_flag)
    
    # Find the three most common errors
    np.fill_diagonal(cm, 0)  # Ignore correct classifications
    errors = np.unravel_index(np.argsort(-cm, axis=None), cm.shape)
    common_errors = list(zip(errors[0][:3], errors[1][:3], cm[errors[0][:3], errors[1][:3]]))

    print("Three most common errors (predicted label, count | true label):")
    for true_label, pred_label, count in common_errors:
        print(f"Predicted label: {letters_no_jz[pred_label]}, Count: {count} | True label: {letters_no_jz[true_label]}")
        
        
def exponential_decay(lr0, s):
    def exponential_decay_fn(epoch):
        return lr0 * 0.1**(epoch / s) 
    return exponential_decay_fn

    
# create list of letters without J and Z
letters_no_jz = np.delete(letters.copy(), [9, 25])
```

### Neural Network

```python
input_shape = X_train.shape[1:]

DenseLayer = partial(layers.Dense, activation='elu', kernel_initializer='he_normal')

model_nn = models.Sequential([
    layers.Input(shape=input_shape),
    DenseLayer(128),
    DenseLayer(128),
    DenseLayer(64),
    DenseLayer(64),
    DenseLayer(32),
    DenseLayer(24, activation='softmax')
])

# Compile the model
model_nn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_nn.summary()
```

```python
history_nn = model_nn.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))
```

```python
y_val_pred = model_nn.predict(X_val)
evaluate_model(y_val, y_val_pred, history_nn, 'Neural Network', plot_cm_flag=False)
```

### Convolutional Neural Network

```python
# prepare the data for CNN
X_train_cnn = X_train.reshape(-1, 32, 32, 1)
X_val_cnn = X_val.reshape(-1, 32, 32, 1)
```

```python
Conv2DLayer = partial(layers.Conv2D, activation='elu', kernel_initializer='he_normal')

model_cnn = models.Sequential([
    Conv2DLayer(32, (3, 3)),
    layers.MaxPooling2D((2, 2)),
    Conv2DLayer(64, (3, 3)),
    layers.MaxPooling2D((2, 2)),
    Conv2DLayer(64, (3, 3)),
    layers.Flatten(),
    layers.Dense(128, activation='elu', kernel_initializer='he_normal'),
    layers.Dense(24, activation='softmax')
])

model_cnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_cnn.summary()
```

```python
y_val_pred_cnn = model_cnn.predict(X_val_cnn)
evaluate_model(y_val, y_val_pred_cnn, history_cnn, 'Convolutional Neural Network', plot_cm_flag=False)
```

# Hyper-Parameter Optimisation

From the two model above, we could see CNN perform much better then the vanilla neural network. Therefore, we will focus on optimising the CNN model. As observed from the learning curve, the model is overfitting. We will try to reduce the overfitting by using dropout and regularisation.

## CNN Model 2

```python
Conv2DLayer = partial(layers.Conv2D, activation='elu', kernel_initializer='he_normal', padding='same')

model_cnn_2 = models.Sequential([
    layers.Input(shape=(32, 32, 1)),
    Conv2DLayer(32, (3,3)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.2),
    Conv2DLayer(64, (3, 3)),
    layers.MaxPooling2D((2, 2)),
    Conv2DLayer(64, (3,3)),
    layers.Flatten(),
    layers.Dense(64, activation='elu', kernel_initializer='he_normal'),
    layers.Dropout(0.2),
    layers.Dense(24, activation='softmax')
])

adam = keras.optimizers.legacy.Adam(learning_rate=1e-4)
model_cnn_2.compile(optimizer=adam,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_cnn_2.summary()
```

```python
history_cnn_2 = model_cnn_2.fit(X_train_cnn, y_train, epochs=20, batch_size=64, validation_data=(X_val_cnn, y_val))
```

```python
y_val_pred_cnn_2 = model_cnn_2.predict(X_val_cnn)
evaluate_model(y_val, y_val_pred_cnn_2, history_cnn_2, 'Convolutional Neural Network 2', plot_cm_flag=False)
```

## CNN Model 3

```python
Conv2DLayer = partial(layers.Conv2D, activation='relu', kernel_initializer='he_normal', padding='same')

model_cnn_3 = models.Sequential([
    layers.Input(shape=(32, 32, 1)),
    Conv2DLayer(128, (5,5)),
    layers.MaxPool2D(pool_size=(3,3),strides=2,padding='same'),
    Conv2DLayer(64, (2,2)),
    layers.MaxPool2D(pool_size=(2,2),strides=2,padding='same'),
    Conv2DLayer(32, (3,3), kernel_regularizer='l2'),
    layers.MaxPool2D(pool_size=(2,2),strides=2,padding='same'),
    layers.Flatten(),
    layers.Dense(512, activation='relu', kernel_initializer='he_normal', kernel_regularizer='l2'),
    layers.Dense(256, activation='relu', kernel_initializer='he_normal', kernel_regularizer='l2'),
    layers.Dense(24, activation='softmax', kernel_initializer="glorot_uniform")
])

adam = keras.optimizers.legacy.Adam(learning_rate=1e-2)
model_cnn_3.compile(optimizer=adam,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_cnn_3.summary()
```

```python
exponential_decay_fn = exponential_decay(lr0=0.01, s=5)
lr_scheduler = keras.callbacks.LearningRateScheduler(exponential_decay_fn)

history_cnn_3 = model_cnn_3.fit(X_train_cnn, y_train, epochs=20, batch_size=64, validation_data=(X_val_cnn, y_val), callbacks=[lr_scheduler])
```

```python
y_val_pred_cnn_3 = model_cnn_3.predict(X_val_cnn)
evaluate_model(y_val, y_val_pred_cnn_3, history_cnn_3, 'Convolutional Neural Network 3', plot_cm_flag=False, show_all_accuracy=True)
```
